### Boxplot / Distribution & Skewness
- Z-Score lebih sesuai digunakan untuk data yang hampir normal atau sedikit skewed.
- Kaedah Tukey (IQR) lebih sesuai untuk data yang skewed secara signifikan dan apabila data mengandungi outlier ekstrem.

- Skewness lebih 3 atau kurang dari -3 = Guna Tukey's Method (IQR).
- Skewness antara -3 dan 3 (sedikit skewed atau normal) = Guna Z-Score.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming numerical_anamolies is a DataFrame with numerical columns
# For example, numerical_anamolies = df.select_dtypes(include=['float64', 'int64'])

# Define the number of columns per row (e.g., 3 columns per row)
n_cols = 3
max_plots_per_figure = 12  # Limit the maximum number of plots per figure

# Calculate how many figures are needed
n_figures = (len(numerical_anamolies) // max_plots_per_figure) + (1 if len(numerical_anamolies) % max_plots_per_figure != 0 else 0)

# Loop through each figure
for fig_idx in range(n_figures):
    # Define the number of rows based on max plots per figure
    n_rows = min(max_plots_per_figure, len(numerical_anamolies) - fig_idx * max_plots_per_figure)

    # Adjust the figure size
    fig_height = n_rows * 4  # 4 inches per row
    fig_width = 14  # Width is kept reasonable at 14 inches

    plt.figure(figsize=(fig_width, fig_height))

    # Loop through each feature for the current figure
    for idx, feature in enumerate(numerical_anamolies.columns[fig_idx * max_plots_per_figure : (fig_idx + 1) * max_plots_per_figure], 1):
        plt.subplot(n_rows, n_cols, idx)  # Use n_rows and n_cols for subplot grid
        sns.histplot(numerical_anamolies[feature], kde=True)
        plt.title(f"{feature} | Skewness: {round(numerical_anamolies[feature].skew(), 2)}")

    # Adjust layout and show plots
    plt.tight_layout()
    plt.show()
```

# PCA 

```python
from sklearn.decomposition import PCA

# Menyediakan data dan melakukan PCA
pca = PCA(n_components=2)  # Merendahkan dimensi kepada 2
pca_result = pca.fit_transform(select_dataset.iloc[:, 1:])

# Menukar ke DataFrame untuk visualisasi
pca_df = pd.DataFrame(pca_result, columns=['PC1', 'PC2'])

sns.relplot(x='PC1', y='PC2', data=pca_df,kind='scatter')
plt.title('PCA - 2D Visualization')
plt.show()
```

# RUN (TRAIN AND TEST THE MODEL)

### K-Means

```python
from sklearn.cluster import KMeans

# Tentukan bilangan cluster, contoh 3 cluster
kmeans = KMeans(n_clusters=2,init='random',random_state=42, max_iter=300)

# Latih model clustering pada data PCA (PC1 dan PC2)
kmeans.fit(pca_df)

# Menyimpan label cluster untuk setiap data
pca_df['Cluster_Kmeans'] = kmeans.labels_

# Visualisasi hasil clustering
plt.figure(figsize=(10, 6))
plt.scatter(pca_df['PC1'], pca_df['PC2'], c=pca_df['Cluster_Kmeans'], cmap='viridis')
plt.title('Clustering using K-Means PCA 2 Dimension')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

### Silhoutte score Kmeans

```python
from sklearn.metrics import silhouette_score

# Kira Silhouette Score
silhouette_avg = silhouette_score(pca_df[['PC1', 'PC2']], pca_df['Cluster_Kmeans'])
print(f'Silhouette Score: {silhouette_avg}')
```

### Elbow Method

```python
# Menentukan jumlah cluster yang berbeza
inertia = []
for i in range(1, 11):  # Uji dari 1 hingga 10 cluster
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(pca_df[['PC1', 'PC2']])
    inertia.append(kmeans.inertia_)

# Plot Elbow Method
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method untuk Menentukan Jumlah Cluster')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Inertia')
plt.show()
```

### DBSCAN

```python
from sklearn.cluster import DBSCAN

# Tentukan parameter DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=3)

# Latih model clustering pada data PCA (PC1 dan PC2)
dbscan.fit(pca_df[['PC1', 'PC2']])

# Menyimpan label cluster untuk setiap data
pca_df['Cluster_dbscan'] = dbscan.labels_

# Visualisasi hasil clustering
plt.figure(figsize=(10, 6))
plt.scatter(pca_df['PC1'], pca_df['PC2'], c=pca_df['Cluster_dbscan'], cmap='viridis')
plt.title('Clustering Using DBSCAN PCA 2D')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

### Silhoutte score DBSCAN

```python
from sklearn.metrics import silhouette_score

# Tapis data untuk hanya menggunakan titik yang tidak dikelaskan sebagai noise (-1)
pca_df_clean = pca_df[pca_df['Cluster_dbscan'] != -1]

# Kira Silhouette Score
silhouette_dbscan = silhouette_score(pca_df_clean[['PC1', 'PC2']], pca_df_clean['Cluster_dbscan'])

print(f"Silhouette Score untuk DBSCAN: {silhouette_dbscan}")
```
